# 최대공약수, 최소공배수 구하기

> 두 수를 입력받아 두 수의 최대공약수와 최소공배수를 반환하는 함수 `gcdlcm`을 작성하세요.
>
> 배열의 맨 앞에 최대공약수, 그 다음 최소공배수를 넣어 반환하세요. 
>
> 예를 들어 두 수 3, 12의 최대공약수는 3, 최소공배수는 12이므로 `gcdlcm(3, 12)`는 `(3, 12)`를 반환해야 합니다. 

---

예시)

```python
print(gcdlcm(3, 12)) #=> (3, 12)
print(gcdlcm(1071, 1029)) #=> (21, 52479)
```

````python
# 아래에 코드를 작성하세요.
````

````python
def gcdlcm(x, y):
    big, small = (x, y) if x > y else (y, x)
        
    # 최대공약수
    for gcd in range(small, 2, -1):
        if small % gcd == 0 and big % gcd == 0:
            break
    
    # 최소공배수
    lcm = big * small / gcd
    
    return gcd, int(lcm)
````

````python
# 해당 코드를 통해 올바른 결과가 나오는지 확인하세요.
print(gcdlcm(3, 12))
print(gcdlcm(1071, 1029))
````

# 불쌍한 달팽이 - 고전 수학문제

> 달팽이는 낮 시간 동안에 기둥을 올라갑니다. 하지만 밤에는 잠을 자면서 어느 정도의 거리만큼 미끄러집니다. (낮 시간 동안 올라간 거리보다는 적게 미끄러집니다.)
> 
> 달팽이가 기둥의 꼭대기에 도달하는 날까지 걸리는 시간을 반환하는 함수를 작성하세요.

> 함수의 인자는 다음과 같습니다.
1. 기둥의 높이(미터)
2. 낮 시간 동안 달팽이가 올라가는 거리(미터)
3. 달팽이가 야간에 잠을 자는 동안 미끄러지는 거리(미터)

---

예시)
```python
snail(100, 5, 2)
# => 33
```

````python
# 아래에 코드를 작성하세요.
````

````python
# 올라가기
def snail(pillar, up, down):
    days = 0
    # n 에 지금까지 올라간 높이를 쌓기
    n = 0
    
    while True:
        days += 1
        n += up
        if n >= pillar:
            return days
        n -= down
    
````

````python
# 해당 코드를 통해 올바른 결과가 나오는지 확인하세요.
print(snail(100, 5, 2))
````

````python
def snail(pillar, up, down):
    days = 0
    
    while 1:
        days += 1
        # 기둥에서 숫자를 빼기
        pillar -= up
        if pillar <= 0:
            return days
        pillar += down
````

````python
# 해당 코드를 통해 올바른 결과가 나오는지 확인하세요.
print(snail(100, 5, 2))
````

# 무엇이 중복일까

> 다음 리스트에서 중복되는 요소만 뽑아서 새로운 리스트를 반환하는 함수를 작성하세요.

---

예시)

```python
duplicated(['a', 'b', 'c', 'b', 'd', 'm', 'n', 'n', 'b'])

['b', 'n']
```

````python
# 아래에 코드를 작성하세요.
````

````python
def duplicated(chars):
    
    # 알파벳의 개수를 셀 dict
    # {'a': 1, 'b': 3, 'c': 1, ...} => 이런 모양으로 완성됨
    char_count = {}
    
    for char in chars:
        # 만약 dict에 이미 해당 알파벳이 key로 있으면
        if char_count.get(char):
            # 해당 알파벳의 등장 회수 +1
            char_count[char] += 1
        # dict에 해당 알파벳이 없으면(처음 등장이면)
        else:
            # 해당 알파벳을 dict에 등록하고 value로 1설정
            char_count[char] = 1
    
    # 답을 모을 리스트
    answer = []
    # k는 알파벳, v는 등장 회수
    for k, v in char_count.items():
        # 등장회수(v)가 1보다 크다면
        if v > 1:
            # 그 알파벳을 리스트에 추가
            answer.append(k)
            
    return answer
````

````python
def duplicated(chars):
    # 한번이라도 등장한 문자열 모으는 list
    registered = []
    answer = []
    
    # 알파벳 순회하며
    for char in chars:
        # 등장한 적 없는 알파벳이면
        if char not in registered:
            # 등장 사실 기록
            registered.append(char)
        # 등장한 적은 있으나, 답에는 없다면
        elif char not in answer:
            # 답에 추가
            answer.append(char)
    
    return answer
````

````python
# 해당 코드를 통해 올바른 결과가 나오는지 확인하세요.
duplicated(['a', 'b', 'c', 'b', 'd', 'm', 'n', 'n', 'b'])
````

````python
def duplicated(chars):
    # 알파벳에서 중복 제거
    uniqs = set(chars)
    
    # 중복 없는 상태에서 하나씩 순회하며
    for uniq in uniqs:
        # 그 알파벳을 지운다. => 한번만 등장한 알파벳 모두 사라짐
        chars.remove(uniq)
        
    # 3번 이상 등장한 알파벳들은 삭제 후에 2개 이상 남아있으므로
    # set으로 중복을 제거 => 다시 list로 변환 => 정렬(굳이 필요없음)
    return sorted(list(set(chars)))
````

````python
# 해당 코드를 통해 올바른 결과가 나오는지 확인하세요.
duplicated(['a', 'b', 'c', 'b', 'd', 'm', 'n', 'n', 'b'])
````

````python
def duplicated(chars):
    check = ''
    
    def check_dup(char):
        # check 가 global에 있지는 않지만
        # 상위 함수에 있을 때는 nonlocal을 사용
        nonlocal check
        if chars.count(char) > 1 and char not in check:
            check += char
            return True
        else:
            check += char
            return False
    
    # 필터 함수를 활용할 수 있다.
    return list(filter(check_dup, chars))
````

````python
# 해당 코드를 통해 올바른 결과가 나오는지 확인하세요.
duplicated(['a', 'b', 'c', 'b', 'd', 'm', 'n', 'n', 'b'])
````

````python
def duplicated(chars):
    # 좋은 코드 아님!
    # 필터 사용 가능 => 람다로 전환 가능을 보기 위함
    return list(set(filter(lambda char: chars.count(char) > 1, chars)))
````

# 알파벳만 남기고 뒤집기

> 문자열이 주어지면, 해당 문자열 중에서 알파벳이 아닌 문자는 전부 빼고 거꾸로 뒤집어 반환하는 함수를 작성하세요.

---

예시)

```python
reverse_letter('krishan')
'nahsirk'

reverse_letter('ultr53o?n')
'nortlu'
```

````python
# 아래에 코드를 작성하세요.
````

````python
def reverse_letter(word):
    answer = ''
    
    for char in word:
        if 'a' <= char <= 'z' or 'A' <= char <= 'Z':
            answer += char
    
    return answer[::-1]
````

````python
def reverse_letter(word):
    answer = ''
    
    for char in word:
        if char.isalpha():
            # 마지막에 뒤집지 않고 처음부터 거꾸로 쌓기
            answer = char + answer
    
    return answer
````

````python
# 해당 코드를 통해 올바른 결과가 나오는지 확인하세요.
print(reverse_letter('krishan'))
print(reverse_letter('ultr53o?n'))
````

# 편-안한 단어

> (QWERTY 키보드를 사용하여 타이핑 한다고 가정할 때) '편안한 단어'는 타이핑 할 때 **손을 번갈아 칠 수 있는 단어**를 말합니다.
>
> 단어를 인자로 받아 그것이 '편안한 단어'인지 여부를 True/False로 반환하는 함수를 작성하세요.
* 모든 단어는 a ~ z까지 오름차순으로 구성된 문자열입니다.

> 문자 목록
- 왼손: q, w, e, r, t, a, s, s, d, f, g, z, x, c, v, b
- 오른손: y, u, i, o, p, h, j, k, l, n, m

````python
# 아래에 코드를 작성하세요.
````

````python
def comfortable_word(word):
    right = 'qwertasdfgzxcvb'
    left = 'yuiophjklnm'
    first_char = word[0]
    
    # 단어의 첫글자가 오른손부터 시작이면,
    if first_char in right:
        # r => l => r => l 순으로 확인 필요
        hands = [right, left]
    
    # 단어의 첫글자가 왼손부터 시작이면,
    elif first_char in left:
        # l => r => l => r 순으로 확인 필요
        hands = [left, right]
    
    
    for idx, char in enumerate(word):
        # 자동으로 l => r / r => l 교체하며 글자 비교
        if char not in hands[idx%2]:
            return False
    
    return True
````

````python
# 해당 코드를 통해 올바른 결과가 나오는지 확인하세요.
print(comfortable_word('qywu'))  # True
print(comfortable_word('apple'))  # False
print(comfortable_word('pqowieur'))  # True
````

# 숫자패턴

>원하는 행까지 아래의 패턴을 생성하는 함수를 작성하세요. 만약 인자가 0이나 음의 정수인 경우 빈 문자열(`''`)을 반환하세요.
> 
> 짝수가 인수로 전달되면 패턴은 통과된 짝수보다 작은 최대 홀수까지 계속되어야 합니다.

---

예시)

```python
print(pattern(9)):

1
333
55555
7777777
999999999

print(pattern(6))
1
333
55555
```
**유의) 패턴에 공백은 없습니다.**

````python
# 아래에 코드를 작성하세요.
````

````python
def pattern(n):
    answer = ''
    
    if n <= 0:
        return answer
    
    # 홀수면 1 증가 => range에 들어가면 다시 1 감소하기 때문
    # 짝수면 그대로 range에 들어가면 1 감소하여 홀수가 됨
    if n % 2:
        n += 1
    
    # 한줄 만들고, 마지막에 엔터(\n) 추가하기
    for i in range(1, n):
        if i % 2:
            answer += f'{i}' * i + '\n'
            
    return answer
    # 만약 맨 마지막 \n을 빼고싶다면? rstrip() => 오른쪽 공백관련 문자열 제거
    # return answer.rstrip()  
````

````python
# 해당 코드를 통해 올바른 결과가 나오는지 확인하세요.
print(pattern(9))
print(pattern(6))
````

# 숫자가 좋아

> 섞여있는 문자열 속에서 정수만 뽑아내 합을 반환하는 함수 `pick_and_sum(words)`를 작성하세요.

---

예시)
```python
pick_and_sum('The30quick20brown10f0x1203jumps914ov3r1349the102l4zy dog') #=> 3635
```

````python
# 아래에 코드를 작성하세요.
````

````python
def pick_and_sum(word):
    total = 0
    
    # 숫자로 이루어진 글자를 모을 빈 문자열
    numstr = ''
    
    for char in word:
        # 숫자로 이루어진 글자면
        if char.isnumeric():
            # numstr에 쌓아나감
            numstr += char
        # 더이상 숫자로 이루어진 글자가 아니고 
        # 기존에 쌓아놓은 numstr가 내용이 있으면 (elif len(numstr) != 0 과 같음)
        elif numstr:
            # 정수로 바꿔서 누적
            total += int(numstr)
            # 문자열 비우고 처음부터
            numstr = ''
            
    return total
    
````

````python
# 해당 코드를 통해 올바른 결과가 나오는지 확인하세요.
print(pick_and_sum('The30quick20brown10f0x1203jumps914ov3r1349the102l4zy dog'))
````
