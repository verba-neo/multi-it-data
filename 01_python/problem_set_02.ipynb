## 구구단 출력하기
> 2단부터 9단까지 반복문을 사용하여 구구단을 출력하세요.
---
**[출력 예시]**
```
------- [2 단] -------
2 X 1 = 2
2 X 2 = 4
2 X 3 = 6
2 X 4 = 8
2 X 5 = 10
2 X 6 = 12
2 X 7 = 14
2 X 8 = 16
2 X 9 = 18

------- [3 단] -------
...
```

````python
# 아래에 코드를 작성하세요.
````

````python
for dan in range(2, 10):
    print(f'------- [{dan} 단] ------- ')
    for num in range(1, 10):
        print(f'{dan} X {num} = {dan*num}')
    print()
````

## 개인정보보호
> 사용자의 핸드폰번호를 입력 받고, 개인정보 보호를 위하여 뒷자리 4자리를 제외하고는 마스킹 처리하세요.
> * 핸드폰번호는 010으로 시작해야하고 11자리여야한다.
> * 핸드폰번호를 입력하지 않았다면 "핸드폰번호를 입력하세요"를 출력한다.

---

**[입력 예시]**

01012341234

**[출력 예시]**

*******1234

````python
# 아래에 코드를 작성하세요.
````

````python
# phone_number = input('폰 번호를 입력해 주세요: ')

phone_number = '01012345678'
secret_number = ''

idx = 0

for idx, char in enumerate(phone_number):
    if idx < 7:
        secret_number += '*'
    else:
        secret_number += char
    
print(secret_number)
````

````python
secret_number = '*' * 7 + phone_number[-4:]
print(secret_number)
````

## 정중앙
> 사용자가 입력한 문자열중 가운데 글자를 출력하세요. 
> * 단, 문자열이 짝수라면 가운데 두글자를 출력하세요.

---

**[입력 예시]**

abc

**[출력 예시]**

b

````python
# 아래에 코드를 작성하세요.
````

````python
string = input('문자열 ㄱㄱ:')

mid = len(string) // 2

if len(string) % 2:
    print(string[mid])
else:
    print(string[mid-1:mid+1])
````

## 소수 찾기

> 조건, 반복문을 응용하여 numbers 리스트의 요소들이 소수인지 아닌지 판단하는 코드를 작성하세요.

---


**[입력 예시]**

26, 39, 51, 53, 57, 79, 85

**[출력 예시]**
```python
26 는 소수가 아닙니다. 2 는 26 의 인수입니다.
39 는 소수가 아닙니다. 3 는 39 의 인수입니다.
51 는 소수가 아닙니다. 3 는 51 의 인수입니다.
53 는 소수입니다.
57 는 소수가 아닙니다. 3 는 57 의 인수입니다.
79 는 소수입니다.
85 는 소수가 아닙니다. 5 는 85 의 인수입니다.
```

````python
numbers = [26, 39, 51, 53, 57, 79, 85]

# 아래에 코드를 작성하세요.
````

````python
for number in numbers:
    
    for insoo in range(2, number):
        if number % insoo == 0:
            print(f'{number} 는 소수가 아닙니다. {insoo}는 {number}의 인수입니다.')
            break
    else:
        print(f'{number} 는 소수입니다.')
````

````python
def check_prime(number):
    for insoo in range(2, number):
        if number % insoo == 0:
            return f'{number} 는 소수가 아닙니다. {insoo}는 {number}의 인수입니다.'
    
    return f'{number} 는 소수입니다.'


numbers = [26, 39, 51, 53, 57, 79, 85]


for num in numbers:
    print(check_prime(num))
````
